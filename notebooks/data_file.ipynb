```python
import pandas as pd
```

```python
# Replace 'path/to/your/textfile.txt' with the actual path to your text document
file_path = 'Project-1-AIPI590/data/raw/medicalCorpus.txt'

# Initialize lists to hold the columns' data
paraphrase_indicators = []
sentence1 = []
sentence2 = []
```

```python
# Open and read the file
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        # print(line[0])
        # Assuming each component is separated by a tab
        parts = line.strip().split('.')  # Adjust the split character if necessary
        parts = [part.strip().replace('\t', ' ') for part in parts if part]
        # print(parts)
        if len(parts) == 2:  # Ensure the line has three parts
            paraphrase_indicators.append(int(parts[0][0]))
            sentence1.append(str(parts[0][2:]))
            sentence2.append(str(parts[1]))
print(sentence1)
print(sentence2)
```

```python
# Create a DataFrame
df = pd.DataFrame({
    'Paraphrase_Indicator': paraphrase_indicators,
    'Sentence_1': sentence1,
    'Sentence_2': sentence2
})

# Display the first few rows of the DataFrame to verify
df.head()
```

```python
# Remove duplicates based on 'Sentence_1' while keeping the first occurrence
df_unique = df.drop_duplicates(subset=['Sentence_1'], keep='first')

# Display the first few rows of the DataFrame to verify the removal
df_unique.head()
```

```python
# If you need to save this DataFrame to a new CSV file
df_unique.to_csv('Project-1-AIPI590/data/processed/medicalCorpus.csv', index=False)
```
